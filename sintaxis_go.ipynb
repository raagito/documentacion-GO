# Introducción a Go
Go (o Golang) es un lenguaje de programación compilado, eficiente y fácil de aprender. Destaca por su concurrencia, rapidez y claridad. Es ideal para sistemas distribuidos, servidores web, herramientas CLI y aplicaciones en la nube.

## Conceptos Básicos
Go es un lenguaje de tipado estático, lo que significa que el tipo de cada variable debe conocerse en tiempo de compilación. Los tipos básicos incluyen enteros, flotantes, booleanos, cadenas, bytes y runas. También existen tipos compuestos como arrays, slices, mapas y structs.

| Categoría | Tipo | Ejemplo |
| --- | --- | --- |
| Enteros | `int16`, `int32`, `uint` | `var edad int16 = 21` |
| Flotantes | `float32`, `float64` | `precio := 12.5` |
| Booleano | `bool` | `var activo bool = false` |
| Cadena | `string` | `mensaje := "Hola"` |
| Byte | `byte` | `var b byte = 'B'` |
| Rune | `rune` | `var r rune = 'Ω'` |
| Array | `[n]tipo` | `var notas [2]int = [2]int{7, 8}` |
| Slice | `[]tipo` | `colores := []string{"rojo", "azul"}` |
| Map | `map[tipo]tipo` | `usuarios := map[string]bool{"Juan": true}` |
| Struct | `type Nombre struct` | `type Libro struct { Titulo string }` |

## Declaración de Variables y Constantes
Las variables pueden declararse usando `var` o el operador corto `:=` (solo dentro de funciones). Si no se asigna un valor, la variable toma el valor cero de su tipo. Las constantes se definen con `const` y su valor debe ser conocido en tiempo de compilación.

```go
var ciudad string = "Madrid"
var cantidad int // valor cero: 0
activo := true // declaración corta
const PI = 3.14
const MaxUsuarios = 200
```

````python
import "fmt"

// Declaración de variables
var pais string = "Argentina"
var edad uint8 = 29
var disponible bool // valor cero: false

// Declaración corta
temperatura := 22.7
ciudad := "Lima"

// Constantes
const Version = "1.0.0"
const DiasSemana = 7

fmt.Println(pais, edad, disponible)
fmt.Println(temperatura, ciudad)
fmt.Println("Versión:", Version, "Días de la semana:", DiasSemana)
````

## Funciones en Go
Las funciones se definen con la palabra clave `func`. Los parámetros y el tipo de retorno deben especificarse. Go permite múltiples valores de retorno y funciones variádicas.

```go
func mostrarMensaje(mensaje string) {
    fmt.Println(mensaje)
}

func multiplicar(a, b int) int {
    return a * b
}

func dividirYModulo(a, b int) (int, int) {
    return a / b, a % b
}

func sumarNumeros(nums ...int) int {
    suma := 0
    for _, n := range nums {
        suma += n
    }
    return suma
}
```

````python
import "fmt"

func mostrarMensaje(mensaje string) {
    fmt.Println("Mensaje:", mensaje)
}

func multiplicar(a, b int) int {
    return a * b
}

func dividirYModulo(a, b int) (int, int) {
    return a / b, a % b
}

func sumarNumeros(nums ...int) int {
    total := 0
    for _, v := range nums {
        total += v
    }
    return total
}

mostrarMensaje("¡Bienvenido a Go!")
fmt.Println("4 x 7 =", multiplicar(4, 7))
cociente, resto := dividirYModulo(20, 3)
fmt.Println("20 / 3 =", cociente, "resto:", resto)
fmt.Println("Suma total:", sumarNumeros(2, 4, 6, 8))
````

## Condicionales y Switch
Go utiliza `if`, `else if` y `else` para tomar decisiones. No se usan paréntesis en la condición, pero las llaves son obligatorias. El `switch` permite evaluar múltiples casos de manera eficiente.

```go
edad := 18
if edad >= 18 {
    fmt.Println("Mayor de edad")
} else {
    fmt.Println("Menor de edad")
}

// Switch
dia := "martes"
switch dia {
case "lunes":
    fmt.Println("Inicio de semana")
case "martes", "miércoles":
    fmt.Println("Mitad de semana")
default:
    fmt.Println("Otro día")
}
```

````python
import "fmt"

edad := 15
if edad >= 21 {
    fmt.Println("Acceso permitido")
} else if edad >= 18 {
    fmt.Println("Acceso restringido")
} else {
    fmt.Println("Acceso denegado")
}

color := "verde"
switch color {
case "rojo":
    fmt.Println("Color de advertencia")
case "amarillo":
    fmt.Println("Color de precaución")
case "verde":
    fmt.Println("Color de seguridad")
default:
    fmt.Println("Color desconocido")
}
````

## Bucles en Go
El único bucle en Go es `for`, pero puede usarse de varias formas: clásico, como `while`, infinito y con `range` para recorrer colecciones.

```go
// for clásico
for i := 1; i <= 3; i++ {
    fmt.Println(i)
}

// for tipo while
contador := 0
for contador < 2 {
    fmt.Println("Contador:", contador)
    contador++
}

// for infinito
for {
    break // salir del bucle
}

// for con range
animales := []string{"gato", "perro"}
for idx, animal := range animales {
    fmt.Println(idx, animal)
}
```

## Punteros en Go
Go permite trabajar con punteros, que almacenan la dirección de memoria de una variable. No existe aritmética de punteros como en C, lo que los hace más seguros. Se usa `&` para obtener la dirección y `*` para acceder al valor referenciado.

```go
numero := 50
puntero := &numero
fmt.Println("Valor:", numero)
fmt.Println("Dirección:", puntero)
fmt.Println("Valor referenciado:", *puntero)
*puntero = 75
fmt.Println("Nuevo valor:", numero)
```

````python
import "fmt"

valor := 12
ptr := &valor
fmt.Println("Valor original:", valor)
fmt.Println("Dirección de memoria:", ptr)
fmt.Println("Valor desde el puntero:", *ptr)
*ptr = 99
fmt.Println("Valor modificado:", valor)
````

## Structs
Los structs permiten agrupar datos relacionados bajo una sola entidad. Son similares a las clases, pero sin herencia. Se pueden asociar métodos a structs usando receivers.

```go
type Animal struct {
    Especie string
    Edad    int
}

func (a Animal) Describir() string {
    return fmt.Sprintf("%s de %d años", a.Especie, a.Edad)
}

perro := Animal{Especie: "Perro", Edad: 4}
fmt.Println(perro.Describir())
```

````python
import "fmt"

type Vehiculo struct {
    Marca string
    Anio  int
}

func (v Vehiculo) Info() string {
    return fmt.Sprintf("%s (%d)", v.Marca, v.Anio)
}

auto := Vehiculo{Marca: "Toyota", Anio: 2020}
fmt.Println(auto.Info())
````

## Interfaces
Las interfaces definen un conjunto de métodos que un tipo debe implementar. En Go, la implementación es implícita: basta con que el tipo tenga los métodos requeridos.

```go
type Sonido interface {
    HacerSonido() string
}

type Gato struct {}

func (g Gato) HacerSonido() string {
    return "Miau"
}

func reproducirSonido(s Sonido) {
    fmt.Println(s.HacerSonido())
}

felix := Gato{}
reproducirSonido(felix)
```

````python
import "fmt"

type Dibujable interface {
    Dibujar() string
}

type Circulo struct {}

func (c Circulo) Dibujar() string {
    return "Dibujando un círculo"
}

func mostrarDibujo(d Dibujable) {
    fmt.Println(d.Dibujar())
}

c := Circulo{}
mostrarDibujo(c)
````

## Manejo de Errores
En Go, los errores se manejan retornando un valor de tipo `error`. Se recomienda siempre verificar si el error es distinto de `nil` tras llamar a una función que puede fallar.

```go
import "errors"

func dividir(a, b int) (int, error) {
    if b == 0 {
        return 0, errors.New("no se puede dividir entre cero")
    }
    return a / b, nil
}

resultado, err := dividir(10, 0)
if err != nil {
    fmt.Println("Error:", err)
} else {
    fmt.Println("Resultado:", resultado)
}
```

````python
import (
    "fmt"
    "errors"
)

func resta(a, b int) (int, error) {
    if b > a {
        return 0, errors.New("el sustraendo es mayor que el minuendo")
    }
    return a - b, nil
}

res, err := resta(3, 8)
if err != nil {
    fmt.Println("Error encontrado:", err)
} else {
    fmt.Println("Resultado:", res)
}
````

## Composición y Polimorfismo
Go no tiene herencia clásica, pero permite la composición de structs (embedding) y el polimorfismo a través de interfaces.

```go
type Motor struct {
    Potencia int
}

type Auto struct {
    Motor
    Marca string
}

miAuto := Auto{Motor: Motor{Potencia: 120}, Marca: "Ford"}
fmt.Println(miAuto.Potencia, miAuto.Marca)
```

````python
import "fmt"

type Bateria struct {
    Capacidad int
}

type Telefono struct {
    Bateria
    Modelo string
}

miTelefono := Telefono{Bateria: Bateria{Capacidad: 4000}, Modelo: "Pixel"}
fmt.Println("Modelo:", miTelefono.Modelo, "mAh:", miTelefono.Capacidad)
````

## Convenciones de Estilo
- Identificadores que empiezan con mayúscula son públicos (exportados), los que empiezan con minúscula son privados.
- Variables locales: `camelCase`. Exportadas: `PascalCase`.
- Paquetes: minúsculas, sin guiones.

| Elemento | Estilo | Ejemplo |
| --- | --- | --- |
| Variable interna | `camelCase` | `totalSuma := 10` |
| Variable exportada | `PascalCase` | `var NumeroMax = 5` |
| Función privada | `camelCase` | `func calcularArea() {}` |
| Función exportada | `PascalCase` | `func CalcularArea() {}` |
| Struct | `PascalCase` | `type Persona struct {}` |
| Interfaz | `PascalCase` + `er` | `type Lector interface {}` |
| Constante | `PascalCase` o `camelCase` | `const TiempoMax = 30` |
| Paquete | minúsculas | `package math` |

## Módulos y Comandos Esenciales
Go gestiona dependencias con módulos. Usa `go mod init` para crear el archivo `go.mod`. Los comandos principales son:

| Comando | Descripción |
| --- | --- |
| `go mod init [nombre]` | Inicializa un módulo |
| `go build` | Compila el proyecto |
| `go run [archivo.go]` | Compila y ejecuta |
| `go get [paquete]` | Añade dependencia |
| `go mod tidy` | Limpia dependencias |
| `go test` | Ejecuta tests |
| `go fmt` | Formatea el código |
| `go vet` | Analiza errores comunes |
